# EDA: Kaggle Datasets (MovieLens + TMDB)

This notebook explores the processed datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Load processed data
ratings = pd.read_parquet('../data/processed/ratings.parquet')
movies = pd.read_parquet('../data/processed/movies.parquet')
```

```python
print(f"Ratings shape: {ratings.shape}")
print(f"Movies shape: {movies.shape}")
```

```python
ratings.head()
```

```python
movies.head()
```

```python
# Distribution of ratings
plt.figure(figsize=(10, 6))
sns.countplot(data=ratings, x='rating')
plt.title('Distribution of Ratings')
plt.show()
```

```python
# Top genres
genres = movies['genres'].str.split('|', expand=True).stack().value_counts()
plt.figure(figsize=(12, 8))
genres.head(20).plot(kind='bar')
plt.title('Top 20 Genres')
plt.show()
```
